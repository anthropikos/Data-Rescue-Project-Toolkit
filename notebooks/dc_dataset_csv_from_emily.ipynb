```python
from arcgis.gis import GIS

def find_match(dataset_name):
    
    print(f"Checking {dataset_name}...")
    
    gis = GIS("https://dcgis.maps.arcgis.com/")
    # items = gis.content.advanced_search(
    items = gis.content.search(
        str(dataset_name), 
        max_items=30, 
        enrich=True,
    )
    
    print(items)
    results = [{"title": item.title, "type": item.type, "homepage": item.homepage, "url": item.url} for item in items]

    print(f"{dataset_name} \t {results}")

    return results

    
find_match("Exposure Activities Number of Activities")
    
```

```python
## Testing if possible to use NLP and similarity metric to match the open data inventory with the open data catalog
##
## No, the similarity metric is matching titles that are not related at all. Whatever feature extraction that Spacy
## is using is not creating a similarity metric as I wanted.

from pathlib import Path
import pandas as pd
import spacy

emily_set_file_path = Path("~/Downloads/OpenDataDC_OpenDataInventory - Open_Data.csv")
public_catelog_file_path = Path("~/Downloads/Open Data DC.csv")

df_em = pd.read_csv(emily_set_file_path)
df_cat = pd.read_csv(public_catelog_file_path)


# Checking dataframe
df_em.DATASET_NAME
df_em.DATASET_DESCRIPTION

df_cat.title
df_cat.description


# Checking dataframe
for idx, data in df_em.iterrows():
    print(idx, data.DATASET_NAME)
    
for row in df_em.itertuples():
    print(row.DATASET_NAME)
    


# Testing similarity metric
nlp = spacy.load("en_core_web_lg")


for idx_em, data_em in df_em.iterrows():
    highest_similarity_score = 0
    idx_of_best_match = None
    title_of_best_match = None

    em_title = data_em.DATASET_NAME
    em_desc = data_em.DATASET_DESCRIPTION

    for idx_cat, data_cat in df_cat.iterrows():        
        
        cat_title = data_cat.title
        cat_desc = data_cat.description

        # print(f"{em_title} \t {cat_title} \t {nlp(em_title).similarity(nlp(cat_title))}")

        sim_score = nlp(em_title).similarity( nlp(cat_title) )
        if sim_score > highest_similarity_score:
            highest_similarity_score = sim_score
            idx_of_best_match = idx_cat
            title_of_best_match = cat_title

    print(f"Best match: \t {em_title} \t {title_of_best_match} \t {highest_similarity_score}")
```
